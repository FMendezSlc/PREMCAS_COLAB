# Modified Hebb-Williams Maze Task

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import pingouin as pg
```

```python
hwk_file = pd.read_csv('/Users/labc02/Documents/PDCB_data/Behavior/HWK_complete.csv')
```

```python
hwk_test = hwk_file[hwk_file['Task Phase'].str.contains('Testing')]
```

```python
gp_ser = [ii.split('_')[0] for ii in hwk_test['ID']]
```

```python
hwk_test['Group'] = gp_ser
```

```python
sns.lineplot(x='Block', y='Error Count', hue='Gene_type', data=hwk_test, estimator='mean', ci=68, style='Group', palette=['forestgreen', 'royalblue'], markers=True, err_style = 'bars', err_kws={'capsize':2, 'capthick':2})
```

```python
pg.mixed_anova(data=hwk_test, dv='Error Count', within='Block', subject='ID', between='Group')
```

```python
pg.pairwise_ttests(data=hwk_test, dv='Error Count', within='Block', between='Group', subject='ID')
```

```python
idx_drop = hwk_test[(hwk_test['Task Phase']=='Testing_4') | (hwk_test['Task Phase']=='Testing_5')].index
```

```python
hwk_class = hwk_test.drop(idx_drop)
```

```python
sns.lineplot(x='Block', y='Error Count', hue='Gene_type', data=hwk_class, estimator='mean', ci=68, style='Group', palette=['forestgreen', 'royalblue'], markers=True, err_style = 'bars', err_kws={'capsize':2, 'capthick':2})
```

```python
groups = ['KOF', 'KOM', 'WTF', 'WTM']
blocks = ['D1_1', 'D1_2', 'D2_1', 'D2_2', 'D3_1', 'D3_2']

for group in groups:
    for block in blocks:
        print(f'{group}, {block}')
        print(pg.normality(hwk_class['Error Count'][(hwk_class['Group'] == group)& (hwk_class['Block']==block)]))
```

```python
pg.homoscedasticity(data=hwk_class, dv='Error Count', group='Group')
```

```python
pg.pairwise_ttests(data=hwk_class, dv='Error Count', between='Group', within='Block', subject='ID', padjust='holm')
```

```python
hw_idx = hwk_class.groupby(['Gene_type', 'Sex', 'ID', 'Block'])['Error Count'].mean()
```

```python
hw_idx = hw_idx.reset_index()
```

```python
hw_idx
```

```python
HW_idx = {'Genotype':[], 'Sex':[], 'ID':[], 'Enc_idx':[], 'Ret_idx':[]}
```

```python
for sub in hw_idx['ID'].unique():
    HW_idx['ID'].append(sub)
    HW_idx['Genotype'].append(hw_idx['Gene_type'][hw_idx['ID']==sub].unique().item())
    HW_idx['Sex'].append(hw_idx['Sex'][hw_idx['ID']==sub].unique().item())
    
    temp_d1_1 = hw_idx['Error Count'][(hw_idx['ID']==sub)&(hw_idx['Block']=='D1_1')].item()
    temp_d1_2 = hw_idx['Error Count'][(hw_idx['ID']==sub)&(hw_idx['Block']=='D1_2')].item()
    temp_d2_1 = hw_idx['Error Count'][(hw_idx['ID']==sub)&(hw_idx['Block']=='D2_1')].item()
    temp_d2_2 = hw_idx['Error Count'][(hw_idx['ID']==sub)&(hw_idx['Block']=='D2_2')].item()
    temp_d3_1 = hw_idx['Error Count'][(hw_idx['ID']==sub)&(hw_idx['Block']=='D3_1')].item()
    temp_d3_2 = hw_idx['Error Count'][(hw_idx['ID']==sub)&(hw_idx['Block']=='D3_2')].item()
    
    
    temp_enc =  ((temp_d1_1-temp_d1_2)+(temp_d2_1-temp_d2_2))/2
    temp_ret =  ((temp_d1_2-temp_d2_1)+(temp_d2_2-temp_d3_1))/2
    
    HW_idx['Enc_idx'].append(temp_enc)
    HW_idx['Ret_idx'].append(temp_ret)
```

```python
K_df = pd.DataFrame(HW_idx)
```

```python
sns.boxplot(data=K_df, x='Genotype', y='Enc_idx')
```
